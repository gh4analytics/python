## Data Cleaning

1. Drop Unwanted Observations
2. Fix Structural Errors
3. Remove unwanted outliers
4. Label missing categorical data
5. Flag and fill missing numerical data

```python
# Load libraries

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
#Load data from csv

df = pd.read_csv('project_files/real_estate_data.csv')
```

## Drop Unwanted Observations

1. Duplicate Observations
2. Irrelevant Observations 

```python
# Drop Duplicates

df = df.drop_duplicates()
df.shape
```

```python
# Irrelevant data
```

## Fix Structural Errors

1. Errors that arise during measurement, data transfer, or other types of "poor housekeeping."

### Wannabe indicator variables

```python
#Unique values for basement

df.basement.unique()
```

```python
#Filling missing (NaN) basement values with 0

df['basement'] = df.basement.fillna(0)
```

```python
df.basement.unique()
```

### Typos and capitalization

```python
sns.countplot(y='roof', data=df)
```

```python
# 'composition' should be 'Composition'
df.roof.replace('composition', 'Composition', inplace=True)

# 'asphalt' should be 'Asphalt'
df.roof.replace('asphalt', 'Asphalt', inplace=True)

# 'shake-shingle' and 'asphalt,shake-shingle' should be 'Shake Shingle'
df.roof.replace(['shake-shingle', 'asphalt,shake-shingle'], 'Shake Shingle', inplace=True)
```

```python
# Class distribution for 'roof'
sns.countplot(y='roof', data=df)
```

### Mislabeled classes

```python
# Class distribution for 'exterior_walls'
sns.countplot(y='exterior_walls', data=df)
```

```python
df.exterior_walls.replace('Rock, Stone', 'Masonry', inplace=True)

df.exterior_walls.replace(['Concrete', 'Block'], 'Concrete Block', inplace=True)
```

```python
sns.countplot(y='exterior_walls', data=df)
```

```python
# Remove lot_size outliers
df = df[df.lot_size < 500000] 

print(len(df))
```

## Label missing categorical data
Handled missing data for categorical features by labeling them as 'Missing'!

```python
# Display number of missing values by feature (categorical)
df.select_dtypes(include=['object']).isnull().sum()
```

```python
df['exterior_walls'] = df['exterior_walls'].fillna('Missing')
```

```python
# Display number of missing values by feature (categorical)
df.select_dtypes(include=['object']).isnull().sum()
```

```python
# Fill missing categorical values
for column in df.select_dtypes(include=['object']):
    df[column] = df[column].fillna('Missing')
```

```python
df.select_dtypes(include=['object']).isnull().sum()
```

## Missing numeric data
Flagging and filling allow the algorithm to estimate the optimal constant for missingness, this technique almost always performs better than imputation methods (for crosssectional data).

```python
# Display number of missing values by feature (numeric)
df.select_dtypes(exclude=['object']).isnull().sum()
```

```python
# Save cleaned dataframe to new file
df.to_csv('project_files/cleaned_df.csv', index=None)
```
